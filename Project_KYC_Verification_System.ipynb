```python
def kyc_check(customer):
    if not customer['pan'] or not customer['aadhaar']:
        return "KYC Incomplete"

    if len(customer['pan']) != 10:
        return "Invalid PAN"

    if len(customer['aadhaar']) != 12:
        return "Invalid Aadhaar"

    if customer['country'] != "India":
        return "High Risk Customer"

    return "KYC Verified"


customer1 = {
    "pan": "ABCDE1234F",
    "aadhaar": "123456789012",
    "country": "India"
}

print(kyc_check(customer1))
```
